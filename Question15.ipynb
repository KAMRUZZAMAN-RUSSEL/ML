```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Sample dataset (Age, CreditScore, RiskClass)
X = np.array([
    [25, 610],
    [30, 640],
    [35, 620],
    [40, 680],
    [45, 700],
    [50, 720]
])

y = np.array([1, 1, 1, 0, 0, 0])  # 1 = Low Risk, 0 = High Risk

# Train SVM
svm_clf = SVC(kernel='linear', C=1e10)
svm_clf.fit(X, y)

# Extract SVM weights
w = svm_clf.coef_[0]
b = svm_clf.intercept_[0]

# Prepare plot
plt.figure(figsize=(8, 6))

# Plot original data points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', s=80, edgecolors='k', label='Data Points')

# Plot SVM decision boundary
x_vals = np.linspace(20, 55, 100)
y_svm = -(w[0] * x_vals + b) / w[1]
plt.plot(x_vals, y_svm, 'k-', label='SVM Hyperplane')

# Plot custom line: y = 12x + 200
y_custom = 12 * x_vals + 200
plt.plot(x_vals, y_custom, 'g--', label='Manual Line: 12x - y + 200 = 0')

# Support vectors
plt.scatter(svm_clf.support_vectors_[:, 0],
            svm_clf.support_vectors_[:, 1],
            s=200, facecolors='none', edgecolors='black', linewidths=1.5, label='Support Vectors')

# Plot formatting
plt.xlabel("Age")
plt.ylabel("Credit Score")
plt.title("SVM vs Manual Hyperplane")
plt.grid(True)
plt.legend()
plt.xlim(20, 55)
plt.ylim(580, 800)
plt.show()
```
